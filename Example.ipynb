# Introduction to RMGP with a toy example

This notebook illustrates the application of the risk model informed Gaussian process (RMGP) framework to update post-earthquake damage estimates using recordings from seismic stations and data from building inspection. 

We load the available data and create a prior risk model. Then we show how an inspection sequence is simulated, and how RMGP updates its components using station data and inspection data. Finally, we also present the implemented random forest and ordered linear probit models that were used in the manuscript for a comparison.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import os
```

The toy example consists of 2000 buildings located in three subregions. For each building, we know its geo-coordinates, construction year, number of storeys and local soil information in terms of $v_{s,30}$. 

We specify the event characteristics and load the station data, which includes the geo-coordinates and the recorded peak ground acceleration (PGA) in g for the three stations.

```python
# Load Building data
dfInput = pd.read_csv('data_toyexample/Buildings.csv')

# Event characteristics
from rmgp.modules import EventChar
EC = EventChar(Magnitude = 6.0, xEpi = -5.0, yEpi = -5.0, style_of_faulting = 'SS')

# Load Station data
dfStation = pd.read_csv('data_toyexample/Stations.csv')
```

## Risk model

Initialize the modules of the prior risk model, which serves as a basis for the RMGP framework. See the manuscript and the supplementary material for further information on this part.

### Ground motion module

```python
from rmgp.gmm_lib.gmms import GMM_AkkarEtAl2014
from rmgp.modules import SpatialGMM

GMM = GMM_AkkarEtAl2014(['x','y','vs30'])

T = 0.0
corr_length = 13.5
SpGMM = SpatialGMM(GMM, corr_length, T)
```

### Damage module

```python
from rmgp.modules import DamageEstimationModule

etas = np.log(np.array([
    [0.15, 0.25, 0.40, 0.60],
    [0.10, 0.20, 0.35, 0.55],
    [0.10, 0.20, 0.30, 0.50],
    [0.30, 0.50, 0.75, 1.30],
    [0.25, 0.45, 0.70, 1.20],
    [0.25, 0.45, 0.65, 1.15],
    [0.70, 1.20, 1.50, 1.90],
    [0.65, 1.15, 1.40, 1.80],
    [0.65, 1.15, 1.35, 1.75]]
    ))
etas[:, 2] = etas[:, 3]
etas = etas[:, :3]
betas = np.array([0.55, 0.65, 0.70, 
                  0.55, 0.70, 0.90,
                  0.55, 0.65, 0.85
                 ]).reshape(-1,1)

DEM = DamageEstimationModule(etas, betas)
```

### Typological Attribution Module

We consider three generic lateral load resisting system types (A, B, C) as the initially unkown attribute combination $a$. The vulnerability types are then obtained by further separating low-/mid-/high-rise buildings.

```python
d_am = {
    'ConPeriod': ['cy $\leq$ 1900', '1900 $<$ cy $\leq$ 1944', '1944 $<$ cy $\leq$ 1969', 
                  '1969 $<$ cy $\leq$ 1988', 'cy $>$ 1988'],
    'p_A': [0.95, 0.60, 0.25, 0.02, 0.01],
    'p_B': [0.05, 0.40, 0.60, 0.25, 0.1],
    'p_C': [0.00, 0.00, 0.15, 0.73, 0.89]
       }
df_am = pd.DataFrame(data=d_am)

# Add a small value to zero entries and normalize such that the rows sum up to one
AttMat = df_am.iloc[:, 1:].values
AttMat[AttMat==0] += 0.001
AttMat = AttMat / np.sum(AttMat, axis=1).reshape(-1,1)

def mean_prior_prob_ta(df_target, AttMat):
    C_A = AttMat.shape[1]
    cy = df_target['const_year'].values
    Xm = np.zeros((len(cy), C_A))
    yb = cy
    for bc in np.arange(C_A):
        Xm[:, bc] = (
            AttMat[0,bc] + (AttMat[1,bc]-AttMat[0,bc]) * 1/(1+np.exp(-0.5*(yb-1900))) +
            (AttMat[2,bc]-AttMat[1,bc]) * 1/(1+np.exp(-0.5*(yb-1944))) +
            (AttMat[3,bc]-AttMat[2,bc]) * 1/(1+np.exp(-0.5*(yb-1969))) +
            (AttMat[4,bc]-AttMat[3,bc]) * 1/(1+np.exp(-0.5*(yb-1988))) )
    return Xm

def AssignVulClass(sim_a, df_target):
    num_story = df_target['num_story'].values
    sim_b = 3 * sim_a
    # Separate height classes according to number of stories
    sim_b += (num_story<=2) * 0 + ((num_story>2) & (num_story<=5)) * 1 + (num_story>5) * 2
    return sim_b.astype('int')

col_names = ['x', 'y', 'const_year', 'num_story']
from rmgp.modules import TypologyAttributionModule
TAM = TypologyAttributionModule(AttMat, 
                          column_names = col_names,
                          column_dtypes = dfInput[col_names].dtypes,
                          func_assign_vul_class = AssignVulClass, 
                          func_prior_mean_prob = mean_prior_prob_ta)
```

```python
from rmgp.model import RMGP
rmgp = RMGP(SpGMM, TAM, DEM, EC)
```

## Inspection sequence

The simulation of an inspection sequence is also described in the manuscript. First, the number of inspection teams that are available in each time step ('AvTeams') is 4. The number of buildings that one team can inspect in one timestep is 5 ('ProdTeam'). We simulate the sequence for 20 time steps ('MaxTimeStep'). 

At the beginning of each time step (inspection day), an un-inspected building is randomly assigned to each available inspection team. After inspecting this building, each inspection team continues by inspecting four additional geographically closest (non-inspected) buildings to complete this time step.

Here we simulate one inspection sequence to illustrate the methodology. For the results presented in the paper we simulated 50 sequences using different seeds. We add the timestep at which each buildings is inspected to the input dataframe. For buildings that were not inspected during the first 20 time steps we assign a time step of 30. Please see also the docstrings of the SimProcess function in the corresponding file.

```python
from utils.sim_insp_sequence import SimProcess
num_seq = 1
# Randomly choose num_seq seeds
rng = np.random.RandomState(seed = 30)
totseeds = rng.choice(np.arange(900), num_seq, replace=False)

InspTimeStep = SimProcess(df=dfInput, seed=totseeds[0], AvTeams=4, ProdTeam=5, MaxTimeStep=20)
dfInput['InspTimeStep'] = InspTimeStep
```

Now we specify the timesteps after which we aim to update the predictions by using inspection data from all buildings that were inspected until this timestep.

```python
pred_timesteps = [2, 4, 6]
```

The figure below shows the locations of all buildings (gray dots) and the locations of the inspected buildings (red crosses) for the three timesteps, where 40, 80 and 120 buildings were inspected.

```python
fig, axs = plt.subplots(1,3, figsize=[9,3], sharex=True, sharey=True)
for i, ax in enumerate(axs):
    mask = dfInput.InspTimeStep <= pred_timesteps[i]
    ax.scatter(dfInput.x, dfInput.y, s=15, color='gray', marker='.', 
               edgecolor='None', alpha=0.7)
    ax.scatter(dfInput[mask].x, dfInput[mask].y, s=20, color='tab:red', marker='x')
    ax.set_aspect('equal'); 
    ax.set_xlabel('x [km]'); 
    if i == 0: ax.set_ylabel('y [km]')
    ax.set_title('m = ' + str(pred_timesteps[i]*20))
plt.show()
```

## Updated predictions

We first do some bookkeeping with the true damage results. The array TrueYzip collects the "true" number of buildings in each damage state for each subregion. This is later used to evaluate the marginal and joint prediction errors.

```python
# Mark inspection results
dfInput.rename(columns={'Type': 'obs_a', 'DS': 'obs_y'}, inplace=True)

# Mark the observed attribute combination as categorical
cols = ['obs_a']
cats = [ ['A', 'B', 'C'] ]
for cat, col in zip(cats, cols):
    cat_var = pd.Categorical(dfInput[col], ordered=True,
                   categories = cat)
    dfInput[col] = cat_var
    
zipI = np.unique(dfInput.subregion)
Nzip = len(zipI)
TrueYzip = np.zeros((4, Nzip))
for i, zipc in enumerate(zipI):
    TrueYzip[:,i] = np.hstack([np.sum(dfInput[(dfInput.subregion==zipc)].obs_y==y) for y in np.arange(4)])
```

Then we initialize two objects which calculate and store statistics (such as the mean and quantiles), as well as the errors (Marginal and Joint prediction errors), based on simulated damage.

```python
from utils.utilsResults import R_stats, R_scores
# Specify the list of quantile which should be evaluated.
qlist = np.array([0.025, 0.05, 0.25, 0.75, 0.95, 0.975])
n_time_steps = len(pred_timesteps) + 1 # Include day 0 (initial risk model pred.)
stats = R_stats(NumClasses=4, n_time_steps=n_time_steps, n_subregions=Nzip, 
                 qlist=qlist)
errors = R_scores(NumClasses=4, n_time_steps=n_time_steps, n_subregions=Nzip)
```

### Updating and prediction workflow

The workflow consists of two loops. First, we perform the updating in each timestep using all data available until this time step. Second, for each subregion we use the updated model to simulate damage. For illustration, we additionally store the posterior mean and variance of the function $f=\ln(im)$, as well as the MAP estimates of the vulnerability function parameters.

```python
# Number of simulations
nSim = 10000
im_list = []
etas_list = []
betas_list = []
for it, t in enumerate(np.append(0,pred_timesteps)):

    # Initialize prior model
    rmgp = RMGP(SpGMM, TAM, DEM, EC)
    # Include station data
    rmgp.Inference_station_data(dfStation, dfStation.PGA.values)
    
    # Include inspection data (if available)
    mask = (dfInput.InspTimeStep <= t)
    if np.sum(mask) > 0:
        obs_a = dfInput[mask].obs_a.values.codes.astype('int32')
        obs_y = dfInput[mask].obs_y.values
        rmgp.Inference_inspection_data(dfInput[mask], obs_a, obs_y)
    
    # Store the mean and variance of the posterior logarithmic im (function f)
    X = rmgp.SpGMM.InitInput(dfInput, EC)
    im_list.append(rmgp.m_im.predict_f(X, full_cov=False))

    # Store the MAP estimates of vulnerability functions parameters
    etas_list.append(rmgp.DEM.etas)
    betas_list.append(rmgp.DEM.betas)
    
    # Loop over subregions
    for k, munic in enumerate(zipI):
        # evaluate number of already inspected buildings in each class in subregion
        y_insp = np.bincount(
            dfInput[(dfInput.subregion==munic) & mask].obs_y.values, minlength=4)  

        # choose remaining buildings in subregion
        dftarget = dfInput[(dfInput.subregion==munic) & (dfInput.InspTimeStep > t)]
        # Sample damage state for each building and return the
        # number of buildings in each damage state for each simulation
        y_sam = rmgp.Sample_damage(dftarget, nSim, aggregate=True)

        # Calculate and allocate statistics (mean and quantiles)
        stats.Allocate(y_samples = y_sam, 
                       y_insp = y_insp, 
                       t_I = it, sr_I = k)
        # Calculate and allocate errors (MPE and JPE)
        errors.Allocate(y_samples = y_sam, 
                        y_insp = y_insp, 
                        y_true = TrueYzip[:, k], 
                        t_I = it, sr_I = k)
```

Below we see how the joint prediction error (JPE) decreases in each subregion (rows) over the considered time steps (columns). 

```python
print('Joint Prediction Error (JPE)')
print(np.round(errors.JPE))
```

This decrease in JPE translates to a decreasing cumulated JPE, as can be seen below. For the results in the paper, we applied the above workflow for 50 random inspection sequences and compared their CJPE values. 

```python
print('Cumulated Joint Prediction Error (CJPE)')
print(np.round(np.sum(errors.JPE,axis=0)))
```

In paralell to the updating of the ground motion field, shown above, RMGP also provides MAP estimates of the vulnerability function parameter. The figure below shows the evolution of $p(Y=1|f)$, where $f$ is expressed as $pga=\exp{f}$. 

```python
from scipy.stats import norm
plot_y = 1 # 0, 1, 2, 3
plot_b = 1 # Vul_class: 0, 1, ..., 8

col = ['black', 'tab:green', 'tab:blue', 'tab:red']
#col = plt.get_cmap('Reds_r')[0, 0.4, 0.6, 0.8]
im = np.linspace(0.001, 1.0, 1000)
f = np.log(im)
fig, ax = plt.subplots(figsize=[5, 3])
for t_s, timestep in enumerate(np.append(0, pred_timesteps)):
    eta_L = np.append(-np.inf, etas_list[t_s][plot_b, :])
    eta_R = np.append(etas_list[t_s][plot_b, :], np.inf)
    beta = betas_list[t_s][plot_b, 0]
    pdf = (norm.cdf((f - eta_L[plot_y])/beta) - 
           norm.cdf((f - eta_R[plot_y])/beta))
    ax.plot(im, pdf, lw=2, color=col[t_s],label='t = '+str(timestep))
ax.legend()
ax.set_ylim(bottom=0)
ax.set_xlabel('Peak ground acc. [g]')
ax.set_title('Conditional prob. of damage state y = ' + str(plot_y))
ax.set_ylabel('$p(y|pga)$')
plt.show()
```

## Purely data driven approaches

In the following, we briefly illustrate the implemenation of the random forest and ordered linear probit model we used in the manuscript.

As of their current implementation, these models do not support the input of a dataframe. Therefore, we first assemple the input matrix X, which consists of all building attributes assumed to be available before the event plus the epicentral distance.

```python
# Assemble Input Matrix
num_col = ['x', 'y', 'const_year', 'num_story', 'vs30']
X_tot = dfInput[num_col].astype('float').values
# Calculate the epicentral distance and include it in
# as an additional feature
R = GMM.get_distance(dfInput, EC).reshape(-1,1)
X_tot = np.append(X_tot, R, axis=1)
```

### Random Forest

We put a wrapper around the scikit implementation of random forests. As explained in the manuscript, this evaluates the out-of-bag score for 12 combinations of 'max_features' and 'min_samples_leaf'. Then it chooses the best performing combination for predictions.

The table below shows the combinations that were used throughout the manuscript.

```python
from other_models.datadriven_models import RF
rf = RF(NumClasses = DEM.c_y)
rf.df_param
```

```python
statsRF = R_stats(NumClasses=4, n_time_steps=4, n_subregions=Nzip, 
                 qlist=qlist)
errorsRF = R_scores(NumClasses=4, n_time_steps=4, n_subregions=Nzip)

for it, t in enumerate(np.append(0,pred_timesteps)):
    # Initialize random forest
    rf = RF(NumClasses = DEM.c_y)
    # Include inspection data
    mask = (dfInput.InspTimeStep <= t)
    if np.sum(mask) == 0: continue
    obs_y = dfInput[mask].obs_y.values
    rf.Inference(X_tot[mask,:], obs_y)

    for k, munic in enumerate(zipI):
        # evaluate number of already inspected buildings in each class in subregion
        y_insp = np.bincount(
            dfInput[(dfInput.subregion==munic) & mask].obs_y.values, minlength=4)
        
        # choose remaining buildings in subregion
        mask_target = ((dfInput.subregion==munic) & (dfInput.InspTimeStep > t))
        y_sam = rf.Sample_damage(X_tot[mask_target, :], nSim, aggregate=True)

        # Calculate and allocate statistics (mean and quantiles)
        statsRF.Allocate(y_samples = y_sam, 
                       y_insp = y_insp, 
                       t_I = it, sr_I = k)
        # Calculate and allocate scores (MPE and JPE)
        errorsRF.Allocate(y_samples = y_sam, 
                        y_insp = y_insp, 
                        y_true = TrueYzip[:, k], 
                        t_I = it, sr_I = k)       
```

### Ordered Linear Probit

We implemented a wrapper around the statsmodel implementation of OrderedModel. For the ordered linear probit model we scale each feature to an identical domain [-1,1].

```python
from sklearn import preprocessing as pp
scaler = pp.MinMaxScaler((-1,1)).fit(X_tot)
X_tot_scaled = scaler.transform(X_tot)
```

```python
from other_models.datadriven_models import OLP
statsOLP = R_stats(NumClasses=4, n_time_steps=4, n_subregions=Nzip, 
                 qlist=qlist)
errorsOLP = R_scores(NumClasses=4, n_time_steps=4, n_subregions=Nzip)

for it, t in enumerate(np.append(0,pred_timesteps)):
    # Initialize ordered linear probit model
    olp = OLP(NumClasses = DEM.c_y)
    # Include inspection data
    mask = (dfInput.InspTimeStep <= t)
    if np.sum(mask) == 0: continue
    obs_y = dfInput[mask].obs_y.values
    olp.Inference(X_tot_scaled[mask,:], obs_y)

    for k, munic in enumerate(zipI):
        # evaluate number of already inspected buildings in each class in subregion
        y_insp = np.bincount(
            dfInput[(dfInput.subregion==munic) & mask].obs_y.values, minlength=4)
        
        # choose remaining buildings in subregion
        mask_target = ((dfInput.subregion==munic) & (dfInput.InspTimeStep > t))
        y_sam = olp.Sample_damage(X_tot_scaled[mask_target, :], nSim, aggregate=True)

        # Calculate and allocate statistics (mean and quantiles)
        statsOLP.Allocate(y_samples = y_sam, 
                       y_insp = y_insp, 
                       t_I = it, sr_I = k)
        # Calculate and allocate scores (MPE and JPE)
        errorsOLP.Allocate(y_samples = y_sam, 
                        y_insp = y_insp, 
                        y_true = TrueYzip[:, k], 
                        t_I = it, sr_I = k)  
```

Now compare the cumulated joint prediction errors for the three methods:

```python
df_res = pd.DataFrame(columns=['Model', 't=0', 't=2', 't=4', 't=6'])
res_list = [np.round(np.sum(errors.JPE,axis=0)),
           np.round(np.sum(errorsRF.JPE,axis=0)),
           np.round(np.sum(errorsOLP.JPE,axis=0))]
models = ['RMGP', 'RF', 'OLP']
for model, cjpe in zip(models, res_list):   
    res = {}
    res = {'Model': model}
    for it, timestep in enumerate(np.append(0,pred_timesteps)):
        string = 't=' + str(timestep)
        res[string] = cjpe[it]
    res = pd.Series(res)
    df_res = pd.concat([df_res,res.to_frame().T],axis=0, ignore_index=True)

df_res
```
